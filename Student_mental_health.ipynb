```python
df=df_raw
```

It looks like the respondents have age between 18 to 24 years.And contains one null value

From above we can say that Timestamp datatype is object and one age is missing, So we have to change the data type of Timestamp and find out the row whose age is nan and replace it with the average of the values of above and bottom


```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'],errors = 'coerce')
```

```python
df.info()
```

```python
#finding out the column number whose age is unknown to us
df['Age'][df['Age'].isna()==True]
```

```python
df.loc[40:45]
```

```python
#replacing nan with average
stu_age_42 = df['Age'][42]
stu_age_44 = df['Age'][44]
stu_age_43 = df['Age'][43]
stu_age_43=(stu_age_42+stu_age_44)/2
```

```python
#rounding of it nearest integer
stu_age_43=stu_age_43.round(0)
```

## Exploratory Analysis and Visualization

Before we ask questions about the responses, it would help to understand the responses.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Counting Courses which are opted by most of the student
course_count=df['What is your course?'].value_counts().reset_index().head()
course_count
```

```python
#Counting number of Female and Male
count_gender=df['Choose your gender'].value_counts().reset_index()
count_gender
```

```python
#Finding Number of Students who are married and not married respectively
stu_marital_status = df['Marital status'].value_counts().reset_index()
stu_marital_status
```

```python
#percentage of students facing depression
facing_deppression = df['Do you have Depression?'][df['Do you have Depression?']=='Yes'].count()

dep_percent= (facing_deppression/df.shape[0])*100
dep_percent
```

```python
#percentage of students facing anxiety
facing_anxiety = df['Do you have Anxiety?'][df['Do you have Anxiety?']=='Yes'].count()
anx_percent = (facing_anxiety/df.shape[0])*100
anx_percent
```

```python
#percentage of student having panic attacks
facing_panic_attack =df['Do you have Panic attack?'][df['Do you have Panic attack?']=='Yes'].count()
panic_attack_percent = (facing_panic_attack/df.shape[0])*100
panic_attack_percent
```

```python
#creating a dataframe of students facing all three problems
students_facingall_df= df[df[['Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?']].nunique(axis=1)==1]
students_facingall_df[students_facingall_df['Do you have Depression?']=='Yes']
```

```python
#Counting students who have consulted any specialist for trreatment or not
df['Did you seek any specialist for a treatment?'].value_counts().reset_index()
```

```python
#Gender Distribution
plt.figure(figsize=(10,8))
plt.hist(df['Choose your gender'],color = 'g')
plt.title('Gender Distribution')
```

```python
#countplot facing Depression and panic attack
plt.figure(figsize=(10,10))
sns.countplot(df['Do you have Depression?'],hue=df['Do you have Panic attack?']);
plt.title("Students facing Depression and Panic attack")
```
